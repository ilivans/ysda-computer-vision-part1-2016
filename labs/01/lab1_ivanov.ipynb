# 4

```python
duckling_mean = duckling[:,:,:3].mean(axis=2)
duckling_mean = (duckling_mean - duckling_mean.min()) / (duckling_mean.max() - duckling_mean.min())
io.imshow(np.concatenate((duckling_mean, color.rgb2gray(duckling)), axis=1))
```

```python
duckling_v = duckling_hsv.copy()
duckling_v[:,:,0] = 0
duckling_v[:,:,1] = 0
io.imshow(duckling_v)
```
